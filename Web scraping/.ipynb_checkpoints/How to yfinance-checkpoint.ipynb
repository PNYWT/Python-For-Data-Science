# Major Indicaies

```python
major_indices = pd.read_html("https://finance.yahoo.com/world-indices/")[0]
```

```python
major_indices.head()
```

```python
major_indices.Symbol
```

```python
major_indices.Symbol = major_indices.Symbol.str.replace(r"\^","")
major_indices.Symbol
```

```python
listSymbol = major_indices.Symbol.to_list()
listSymbol
```

```python
listSymbolDate = yf.Tickers(listSymbol)
listSymbolDate
```

```python
len(listSymbol)
```

```python
dfSymbol = yf.download(listSymbol, period="1d", start="2020-01-01", end="2023-11-08")
```

```python
dfSymbol.columns
```

## Clean Data dfSymbol

```python
dfSymbol.head()
```

```python
adj_close = dfSymbol.dropna(thresh = 10, axis=True)["Adj Close"]
adj_close
```

```python
adj_close.describe().T
```

```python
adj_close.plot(figsize = (6,6))
```

```python
adj_close.plot(figsize = (8,8), subplots = True)
```

### Drop XAX DJI HSI NYA to lack of viable data.

```python
adj_close.plot(figsize = (6,6))
```
