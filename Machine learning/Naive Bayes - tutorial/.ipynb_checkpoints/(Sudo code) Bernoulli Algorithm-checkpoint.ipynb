```python
def calculate_prior_probs(labels): #P(c|X)
    classes = set(labels)
    prior_probs = {}
    total_count = len(labels)
    
    for cls in classes:
        prior_probs[cls] = labels.count(cls) / total_count
    
    return prior_probs
```

```python
def calculate_cond_probs(data, labels, features): #P(x1|C)*P(x2|C)*...*(Pxn|C)*P(c)
    classes = list(set(labels)) #  Class   array 
    num_features = len(features) #  feature 
    cond_probs = {} # (conditional probabilities)  feature  class
    
    for f in range(num_features): # Loop  feature
        feature_counts = [0] * len(classes) # (list)  class
        for i, example in enumerate(data): # Data 
            if example[f] == features[f]: # check  feature  data 
                feature_counts[classes.index(labels[i])] += 1 #  class  + 1
        for cls in classes:  # Class  
            # probability  feature  Class  cond_probs
            cond_probs[(f, cls)] = (feature_counts[classes.index(cls)] + 1) / (labels.count(cls) + 1 + len(features))
    print("index {}, cond_probs {}".format(f,cond_probs))
    return cond_probs
```
