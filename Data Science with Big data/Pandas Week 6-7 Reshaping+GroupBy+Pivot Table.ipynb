## Groupby ในการ split กลุ่มข้อมูลตามค่าที่เป็นไปได้ทั้งหมดใน Category Column

```python
df_player.head()
```

```python
group = df_player.groupby("position")
group
```

```python
group.groups
```

```python
#คำนวน mean passes ของแต่ละกลุ่ม แสดงผลเป็นกลุ่ม position
group.passes.mean()
```

```python
#คำนวน sum passes ของแต่ละกลุ่ม แสดงผลเป็นกลุ่ม team
df_player.groupby("team").passes.sum()
```

## Pivot / melt

```python
path_TempDf = "https://raw.githubusercontent.com/PNYWT/Python-For-Data-Science/main/Data%20Science%20with%20Big%20data/Datasets/eu_city_temp.csv"
temp_df = pd.read_csv(path_TempDf,header=1) #<- ตัด row 1 ทิ้ง
temp_df.head()
```

```python
temp_df = temp_df.iloc[:,:-2]
temp_df.head()
```

```python
#ปรับแนวนอนเป็นแนวตั้ง
temp_df.melt(id_vars=["Country","City"]).head()
```

```python
temp_Long = temp_df.melt(id_vars=["Country","City"],var_name="Month", value_name="teamperature")
temp_Long.head()
```

```python
#ปรับแนวตั้งเป็นแนวนอน
temp_wide = temp_Long.pivot(index=["Country", "City"], columns="Month", values="teamperature")
temp_wide.head()
```
