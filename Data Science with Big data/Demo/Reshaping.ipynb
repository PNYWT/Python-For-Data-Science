## example: calculate % of minutes played by team

```python
df1 = players_df.loc[: , ['surname','team','minutes']]
df1
```

```python
df2 = teams_df.loc[: , ['team', 'games']]
df2
```

```python
df2['total_mins'] = df2.games * 90
df2
```

```python
df3 = pd.merge(df1, df2, on='team', how='outer')
df3
```

```python
df3['percent_played'] = df3.minutes / df3.total_mins * 100
df3
```

```python
df3.query('percent_played > 100')
```
