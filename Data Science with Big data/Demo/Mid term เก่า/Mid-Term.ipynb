# 25. จากตาราง star_7 แสดงแผนภาพแบบ scatterplot แสดงจำนวนครั้งที่เครื่องบินตกเทียบกับจำนวนคนตายเฉลี่ยต่ออุบัติเหตุหนึ่งครั้ง และแยกสีตามสายการบิน ขนาดของจุดแปรผันตามจำนวนคนตายรวมของสายการบิน

```python
starOp_Size = star_7.groupby("Operator").Fatalities.size()
starOp_Size
```

```python
starOp_mean = star_7.groupby("Operator").Fatalities.mean()
starOp_mean
```

```python
starOp_sum = star_7.groupby("Operator").Fatalities.sum()
starOp_sum
```

```python
scatter_df = pd.concat([starOp_Size, starOp_mean, starOp_sum], axis="columns")
scatter_df
```

```python
scatter_df.columns = ["Accident Count", "Average Fatalities", "Total Fatalities"]
scatter_df
```

```python
scatter_df = scatter_df.reset_index()
scatter_df
```

```python
sns.relplot(data = scatter_df, x = "Operator", y = "Average Fatalities",
                hue = "Operator", size = "Total Fatalities", height = 8)
```
