## row/col selection

### iloc 

```python
df.head()
```

```python
df.iloc[0 , 2]
```

```python
df.iloc[0, :] # select all columns of row 0 (Series)
```

```python
df.iloc[ [4,1,3] , : ]
```

```python
df.iloc[ [4,1,3] , [2,1,0]]
```

```python
df.iloc[ 3:10:2 , :8 ]
```

```python
df = df.iloc[ : , :8]
df
```

### set_index

```python
df.set_index('surname')
```

```python
df = df.set_index('surname')
df
```

### loc

```python
df.loc[ 'Abdoun', 'position' ]
```

```python
df.loc[ ['Abdoun', 'Messi', 'Ronaldo'], ['position','team'] ]
```

```python
df.loc[ 'Abdoun':'Chaouchi' , : ]
```

```python
df.loc[ 'Abdoun':'Chaouchi' , 'team':'shots' ]
```

## filter

```python
df.shots
```

```python
df[ df.shots == 0 ]
```

```python
df[ df.shots >= 20 ]
```

```python
df[ df.team == 'Spain' ]
```

```python
'Thailand'.endswith('land')
```

```python
df[ df.team.str.endswith('land') ]
```

```python
'h' in 'Thailand'
```

```python
df[ df.team.str.contains('w') ]
```

```python
df[ df.index.str.contains('v') ]
```

### query (sql-like)

```python
df.query('shots >= 20')
```

### multiple conditions

```python
df[ (df.shots > 10) & (df.minutes <= 270) ]
```

```python
df[ (df.team == 'England') | (df.team == 'Spain') ]
```

```python
df.loc[ df.shots >= 20, ['team', 'position', 'shots'] ]
```

```python
df.query('shots >= 20').loc[:, ['team', 'position', 'shots']]
```
