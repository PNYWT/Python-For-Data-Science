### 19จำนวนเหรียญทั้งหมดที่แจกไปในโอลิมปิกฤดูหนาวรวมแล้วมีกี่เหรียญ

### 20หาค่าเฉลี่ยของจำนวนเหรียญทองฤดูหนาวและฤดูร้อนต่อประเทศ

```python
thisCase = medals[["Country","SummerGold","WinterGold"]]
thisCase = thisCase.set_index("Country")
thisCase.mean()
#medals.mean()
```

### 21แสดงจำนวนเหรียญทั้งหมด(รวมฤดูร้อนและฤดูหนาว)ที่แต่ละประเทศได้

```python
medals.head()
```

```python
medals = medals.set_index("Country")
medals.head()
```

```python
medals["TotalCoinSummer"]  = medals['SummerGold'] + medals['SummerSilver'] + medals['SummerBronze']
```

```python
medals["TotalCoinWinter"] = medals['WinterGold'] + medals['WinterSilver'] + medals['WinterBronze']
medals.head()
```

```python
thisCase = medals[["TotalCoinSummer","TotalCoinWinter"]]
thisCase
```

### 22แสดงประเทศที่ได้เหรียญรวมทุกประเภทของโอลิมปิกฤดูหนาวมากกว่าเหรียญรวมโอลิมปิกฤดูร้อนให้คอลัมน์ของตารางนี้แสดงชื่อประเทศเหรียญรวมฤดูร้อนเหรียญรวมฤดูหนาว

```python
medals.loc[(thisCase.TotalCoinWinter > thisCase.TotalCoinSummer), ["TotalCoinWinter","TotalCoinSummer"]]
```

### 23แสดงประเทศที่ไม่เคยได้เหรียญใดของโอลิมปิกฤดูร้อนแต่ได้เหรียญใดของฤดูหนาวหรือได้แต่เหรียญฤดูร้อนไม่เคยได้ฤดูหนาว

```python
#ไม่เคยได้เหรียญใดของโอลิมปิกฤดูร้อนแต่ได้เหรียญใดของฤดูหนาว
medals.loc[(thisCase.TotalCoinSummer == 0) & (thisCase.TotalCoinWinter >= 1)
              #ได้แต่เหรียญฤดูร้อนไม่เคยได้ฤดูหนาว
             | (thisCase.TotalCoinSummer >= 1) & (thisCase.TotalCoinWinter == 0),["TotalCoinSummer","TotalCoinWinter"]]
```

### 25แสดงประเทศเรียงลำดับตามเหรียญทองฤดูร้อนจากมากไปน้อยถ้าได้เหรียญทองฤดูร้อนเท่ากันให้เรียงตามจำนวนการเข้าแข่งฤดูร้อนจากน้อยไปมาก

```python
# false จากมากไปน้อย, True จากน้อยไปมาก
medals.sort_values(["TotalCoinSummer","SummerGame"], ascending=[False, True])
```

### 26ให้สร้างคอลัมน์ใหม่ในตาราง medals ให้ชื่อว่า score โดยคำนวณคะแนนรวมจากเหรียญที่แต่ละประเทศได้(ทั้งฤดูร้อนและหนาว) โดยให้เหรียญทองได้ 3 คะแนนเหรียญเงินได้ 2 คะแนนเหรียญทองแดงได้ 1 คะแนนแล้วแสดงข้อมูล 10 ประเทศแรกที่ได้คะแนนเยอะที่สุด

```python
medals.head()
```

```python
medals["Score"] = medals['SummerGold']*3 + medals['SummerSilver']*2 + medals['SummerBronze']*1 
+ medals['WinterGold']*3 + medals["WinterSilver"]*2 + medals['WinterBronze']*1  
```

```python
medals.sort_values("Score", ascending=False).head(10)
```

### 27การเปรียบเทียบคะแนนจากเหรียญทั้งหมดอาจจะไม่แฟร์เพราะแต่ละประเทศเข้าร่วมการแข่งขันไม่เท่ากันดังนั้นให้สร้างคอลัมน์ใหม่ในตาราง medals ชื่อว่า perGame โดยคำนวณคะแนนต่อการเข้าร่วมแข่งขัน(คะแนนคำนวณเหมือนเดิมแต่หารด้วยจำนวนเกมส์ที่เข้าแข่งขันแยกกันระหว่างฤดูร้อนกับฤดูหนาวค่าperGame จะเป็นผลบวกของคะแนนของฤดูร้อนกับฤดูหนาว) แสดงข้อมูล 10 ประเทศที่ได้คะแนนน้อยที่สุด

```python
medals.head()
```

```python
medals["perGame"] = (medals['SummerGold']*3 + medals['SummerSilver']*2 + medals['SummerBronze']*1)/ medals["SummerGame"]
+ (medals['WinterGold']*3 + medals["WinterSilver"]*2 + medals['WinterBronze']*1)/ medals["WinterGame"]
```

```python
medals.perGame.fillna(0)
```

```python
medals[["Code","perGame"]].sort_values("perGame", ascending=True).head(10)
```
