## concat ต่อตาราง เหมาะกับตารางที่มี Column คล้ายกัน
pd จะต่อจากชื่อ column ที่ตรงกัน
หรือ เลือกต่อจาก row ที่ตรงกัน ตามแต่กำหนด<br>

pandas.concat(objs, *, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=None)[source]<br>

axis -> 0 = index, 1 = columns, default 0<br>
join -> ‘inner’ ต่อแค่มีร่วมกัน, ‘outer’ ต่อหมดเลยอันไหนไม่มีค่าเป็น NaN, default outer<br>

```python
df_top = df_player.head(5)
df_top
```

```python
df_bottom = df_player.tail(5)
df_bottom
```

```python
df_top = df_top.drop(["minutes","shots"], axis=1)
df_top
```

```python
df_bottom = df_bottom.loc[:,["team","surname","passes","minutes"]]
df_bottom
```

### Top-Bottom ต่อ Table 

```python
pd.concat([df_top,df_bottom])
```

#### แบบระบุ Join ประเภท inner
ต่อเฉพาะ Column ที่ตรงกัน

```python
pd.concat([df_top,df_bottom], join="inner")
```

### Left-Rgiht  ต่อ Table, axis = 1

```python
pd.concat([df_top,df_bottom], axis=1)
```

```python
df_Left = df_player.sample(20)
df_Left
```

```python
df_Right = df_teams.sample(20)
df_Right
```

```python
# drop ประเทศที่ซ้ำทิ้งไป
df_Left.drop_duplicates("team").set_index("team")
```

```python
df_Right.set_index("team")
```

```python
pd.concat([df_Left.drop_duplicates("team").set_index("team"),df_Right.set_index("team")],
          join="outer", axis=1)
```

```python
pd.concat([df_Left.drop_duplicates("team").set_index("team"),df_Right.set_index("team")],
          join="inner", axis=1)
```

### Merge
DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=None, indicator=False, validate=None)<br>
Left Join, Right Join <br>
param -> how{‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}, default ‘inner’}

### รวมตารางเข้าด้วยกัน

```python
# merge column ที่มีค่า team ตรงกัน
pd.merge(df_Left, df_Right, how="inner", on="team")
```

```python
# ยึดตาราง left เป็นหลัก อันไหนไม่มีค่า กำหนดให้เป็น NaN
pd.merge(df_Left, df_Right, how="left", on="team")
```

```python
# ยึดตาราง right เป็นหลัก อันไหนไม่มีค่า กำหนดให้เป็น NaN
pd.merge(df_Left, df_Right, how="right", on="team")
```

# Example Calucalte % of played by team

```python
dfExPlayer = df_player.loc[:,["surname","team","minutes"]]
dfExPlayer
```

```python
dfExTeam = df_teams.loc[:,["team","games"]]
dfExTeam
```

```python
dfExTeam["totalTime"] = dfExTeam.games*90
dfExTeam
```

```python
dfMerge = pd.merge(dfExPlayer, dfExTeam, how="outer", on="team")
dfMerge
```

```python
dfMerge.minutes / dfMerge.totalTime
```

```python
dfMerge["Percent_played"] = (dfMerge.minutes / dfMerge.totalTime)*100
dfMerge
```

```python
dfMerge.loc[dfMerge.Percent_played > 100].head()
```
