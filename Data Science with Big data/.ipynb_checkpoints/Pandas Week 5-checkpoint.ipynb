```python
path = "https://raw.githubusercontent.com/PNYWT/Python-For-Data-Science/main/Data%20Science%20with%20Big%20data/Datasets/players.csv"
df = pd.read_csv(path)
```

```python
df = df.loc[:,:"saves"]
```

### Sort Data

```python
#ascending=False มากไปน้อย
#ascending=True น้อยไปมาก
df.sort_values("passes", ascending=False)
```
