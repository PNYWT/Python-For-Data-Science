```python
import pandas as pd
```

```python
path = "Datasets/players.csv"
player_df = pd.read_csv(path)
player_df.head()
```

```python
player_df = player_df.loc[:,:"saves"]
player_df.head()
```

## สร้าง Categorical

```python
player_df.position.astype("category")
```

```python
pd.Categorical(player_df.position)
```

```python
player_df["position_Category"] = pd.Categorical(player_df.position)
player_df.head()
```

##  เรียงลำดับ Category

```python
pd.Categorical(player_df.position, ordered=True)
```

```python
player_df["position_Category"] = pd.Categorical(player_df.position
                                                , categories = ["goalkeeper", "defender", "midfielder", "forward"]
                                                , ordered=True)
player_df.head()
```

```python
player_df.sort_values(["team","position"]).head()
```

```python
player_df.loc[player_df.team == "Algeria", :].sort_values(["team","position"]) #sort string
```

```python
player_df.loc[player_df.team == "Algeria", :].sort_values("position_Category") #sort Category
```

## การจัดการกบั Category

### เรียกดูชื่อ Categories 

```python
player_df.position_Category.cat.categories
```

### Remove Categories

```python
player_df.position_Category.cat.remove_categories("goalkeeper")
```

### Add Categories

```python
player_df.position_Category.cat.add_categories("coach")
```

```python
s_cat = player_df.position_Category.cat.add_categories("coach")
s_cat
```

```python
s_cat[0] = "coach" # index 0 midfielder -> coach
s_cat
```

```python
s_cat[0] = "midfielder"
s_cat
```

### ยกเลิกลำดับ Catagory กลับเป็นไปเป็น String 

```python
s_cat.cat.as_unordered()
```

### เรียงลำดับใหม่

```python
s_cat.cat.reorder_categories(["forward","midfielder","defender","goalkeeper","coach"])
```
