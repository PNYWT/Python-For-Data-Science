## Resample 
# เปลี่ยน Index ให้เป็น Timestamp (เหมือน Groupby แต่ในแง้ของเวลา โดยแบ่งตาม Datetime) 

```python
house_df.resample('M').size()
#2014 เดือน 5 มีตาราง 1768 แถว (ขายบ้านไปได้ 1768 หลัง)
#สังเกตุ Defaut timestame จะแสดงวันสุดท้ายของเดือน 
```

```python
house_df.resample('M').size().to_period()
#สังเกตุ Defaut timestame จะแสดงวันสุดท้ายของเดือน > ปรับใหม่ จะแสดงเป็นเดือน แทน
```

```python
house_df.resample('Y').price.mean().to_period()
#ราคาบ้านแต่ละปี แสดงค่าเฉลี่ย
```

```python
house_df.resample('Q').price.mean().to_period()
#ราคาบ้านแต่ละชั่วคอเตอร์ แสดงค่าเฉลี่ย
```

```python
house_df.resample('D').size()
```

```python
house_df.resample('3D').size()
#กรุ๊ปข้อมูล หรือจัดกลุ่ม ทุก ๆ 3 วัน
```

# Shift / diff

```python
#วิธีการคำนวณ เชิงเวลาที่สำคัญ อีกแบบ 
```

```python
daily_sale = house_df.resample('D').size()
daily_sale
```

```python
#หากเราต้องการนำข้อมูลของแต่ละวัน มาคำนวณวิธีการ (เอาค่าของแถวปัจจุบัน มาลบกับเข้ามูลแถวก่อนหน้า)
```

```python
daily_sale.diff(1) #current row - previous row (ผลเฉลยของ คำสั่งที่ 62)
```

```python
daily_sale.diff(7).head(20) #เอาแถวปัจจุบัน ลบ 7 แถวก่อนหน้า
```

```python
daily_sale.diff(-1)
```

```python
#การเลื่อนแถว ขึ้น หรือ ลง
daily_sale
```

```python
daily_sale.shift(1) #สังเกต แถวถูกขยับลงมา 
```

```python
df3 = pd.concat([daily_sale, daily_sale.shift(1)], axis=1)
df3.columns  = ['today_sale','yesterday_dale']
df3 #ข้อมูลแยก 2 columns แล้ว เราเอามาคำนวณได้เช่นกัน
```

```python
df3.today_sale - df3.yesterday_dale
```

```python
df3.today_sale / df3.yesterday_dale
```

## Rolling

```python
daily_sale
```

```python
daily_sale.rolling(3).mean() 
#   2014-05-02    67 + 2014-05-03     4 + 2014-05-04    5 / 3 = 25.333333
#windows มี Length: 391
```

```python
daily_sale.resample('3D').mean()  
#windows ต่างกัน มี Length: 131
#เอาข้อมูลของ 3 วัน คำนวณ 
```

```python
df4 = house_df.price.loc['2014-05-03':'2014-05-10'].sort_index()
df4
```

```python
df4.rolling('2D').max()
```

```python
df4.rolling('2D').max().head(20)
```

```python
df4.rolling('2D').max().groupby('date').last()
#ราคาบ้านที่แพงที่สุดในช่วง 2 วันนี้ 
```
