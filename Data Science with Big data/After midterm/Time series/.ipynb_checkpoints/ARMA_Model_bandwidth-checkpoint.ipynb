```python
import pandas as pd
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
```

```python
bandwidth = pd.read_csv("bandwidth.csv")
bandwidth
```

```python
bandwidth.plot()
```

```python
bw_train = bandwidth[:9000]
bw_test = bandwidth[9000:]
```

## Check Stationarity

```python
adfuller(bw_train) #p-value ถ้าต่ำกว่า 0.5 เป็น Stationarity แล้ว
```

```python
bw_train_diff = bw_train.diff().dropna()
bw_train_diff
```

```python
adfuller(bw_train_diff) #d=1 <- diff แล้ว เป็น Stationarity
```

## Check MA 

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(bw_train_diff)
```

```python
plot_pacf(bw_train_diff, method="yw")
```

## AIC Grid search for p and q

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
results = []
pStop = 5 # stop5 คือ p = 4 
for p in range(pStop):
    for q in range(pStop):
        model = SARIMAX(bw_train, order=(p,1,q))
        fitted_model = model.fit()
        results.append([(p,q), fitted_model.aic])
```

```python
results_df = pd.DataFrame(results, columns=["p,q", "AIC"])
results_df.sort_values("AIC")
```

## fit ARMA

```python
model_arma = SARIMAX(bw_train, order=(2,1,2)) #p,d,q
```

```python
fitted_arma = model_arma.fit()
```

## Residual Analysis for AIC using QQplot

```python
residual = fitted_arma.resid
residual
```

```python
from statsmodels.graphics.gofplots import qqplot
```

```python
qqplot(residual[1:], line="45") #ตามทฤษฏีควรจะเป็นเส้นตรง เรียกว่า white noise
```

## Further analysis

```python
fitted_arma.plot_diagnostics()
```

## Ljung-Box Test

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
acorr_ljungbox(residual, lags=10) #ไม่มี lag ที่ p-value น้อยกว่า 0.05 ดังนั้นเป็น independent
```
