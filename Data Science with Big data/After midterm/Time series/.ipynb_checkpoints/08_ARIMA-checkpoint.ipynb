```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../DataSets/jj.csv')
df
```

```python
df['date'] = pd.to_datetime(df.date)
df = df.set_index('date')
df.columns = ['EPS']
df
```

```python
sns.relplot(df, x=df.index, y='EPS', kind='line', 
            height=3.5, aspect=1.5)
```

```python
train = df.iloc[:-4, :]
test = df.iloc[-4: , :]
test
```

# Baseline

## historical mean 

```python
hist_mean = train.EPS.mean()
hist_mean
```

```python
baselines = test.copy()
baselines.columns = ['actual']
baselines
```

```python
baselines['hist_mean'] = hist_mean
baselines
```

## last value

```python
last_value = train.EPS[-1]
last_value
```

```python
baselines['last_value'] = last_value
baselines
```

## last season mean

```python
last_season = train.EPS[-4:]
last_season
```

```python
last_season.mean()
```

```python
baselines['last_season_mean'] = last_season.mean()
baselines
```

```python
list(last_season)
```

```python
baselines['last_season'] = list(last_season)
baselines
```

# ARMA

```python
bw_df = pd.read_csv('../DataSets/bandwidth.csv')
bw_df
```

```python
bw_df.plot()
```

```python
bw_train = bw_df[:9000]
bw_test = bw_df[9000:]
```

```python
bw_train
```

## check stationarity

```python
adfuller(bw_train)
```

```python
bw_train_diff = bw_train.diff().dropna()
```

```python
adfuller(bw_train_diff)
```

## check MA and AR process

```python
plot_acf(bw_train_diff)
```

```python
plot_pacf(bw_train_diff)
```

```python
results = []
for p in range(5):
    for q in range(5):
        model = SARIMAX(bw_train, order=(p,1,q))
        fitted_model = model.fit()
        aic = fitted_model.aic
        results.append( [p,q,aic] )
```

```python
results
```

```python
results_df = pd.DataFrame(results, columns=['p','q','aic'])
results_df
```

```python
results_df.sort_values('aic')
```

```python
model_arma = SARIMAX(bw_train, order=(2,1,2))
fitted_arma = model_arma.fit()
```

## residual analysis

```python
fitted_arma.resid
```

```python
from statsmodels.graphics.gofplots import qqplot
```

```python
qqplot(fitted_arma.resid[1:], line='45')
```

```python
fitted_arma.plot_diagnostics()
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
acorr_ljungbox(fitted_arma.resid, lags=10)
```

# Seasonality

```python
air_passenger = pd.read_csv('../DataSets/air-passengers.csv')
```

```python
air_passenger.plot()
```

```python
air_passenger
```

```python
air_passenger_train = air_passenger.iloc[:-12, 1]
air_passenger_test = air_passenger.iloc[-12:, 1]
```

```python
air_passenger_train
```

## check stationarity

```python
adfuller(air_passenger_train)
```

```python
adfuller(air_passenger_train.diff().dropna())
```

```python
adfuller(air_passenger_train.diff(12).dropna())
```

```python
adfuller(air_passenger_train.diff().diff().dropna())
```

```python
adfuller(air_passenger_train.diff().diff(12).dropna())
```

```python
air_passenger_train_diff = air_passenger_train.diff().diff().dropna()
```

```python
plot_acf(air_passenger_train.diff().dropna(), lags=30)
```

```python
results = []
for p in range(1,3):
    for q in range(1,3):
        for P in range(1,3):
            for Q in range(1,3):
                model = SARIMAX(air_passenger_train,
                                order=(p,1,q),
                                seasonal_order=(P,1,Q,12))
                fitted_sarima = model.fit()
                results.append([p,q,P,Q, fitted_sarima.aic])
```

```python
results_df = pd.DataFrame(results, columns=['p','q','P','Q','AIC'])
results_df
```

```python
results_df.sort_values('AIC')
```

```python
fitted_sarima = SARIMAX(air_passenger_train,
                        order=(2,1,1), 
                        seasonal_order=(1,1,2,12)
                       ).fit()
```

```python
fitted_sarima.plot_diagnostics()
```

```python
acorr_ljungbox(fitted_sarima.resid)
```

# SARIMAX

```python
import statsmodels.api as sm
```

```python
gdp_df = sm.datasets.macrodata.load_pandas().data
gdp_df
```

```python
gdp_df.realgdp.plot()
```

```python
gdp_train = gdp_df.iloc[:-20,2:8]
gdp_test = gdp_df.iloc[-20:,2:8]
```

```python
gdp_train
```

```python
realgdp_train = gdp_train.realgdp
features_train = gdp_train.iloc[:, 1:]
realgdp_test = gdp_test.realgdp
features_test = gdp_test.iloc[:,1:]
```

```python
features_train
```

```python
fitted_sarimax = SARIMAX(realgdp_train,
                         features_train,
                         order=(3,1,3),
                         seasonal_order=(0,0,0,4)).fit()
```

```python
predicted_gdp = fitted_sarimax.forecast(steps=20, exog=features_test)
predicted_gdp
```

```python
pd.concat([predicted_gdp, realgdp_test], axis=1).plot()
```

```python
fitted_sarimax.summary()
```

```python
fitted_sarimax.plot_diagnostics()
```
