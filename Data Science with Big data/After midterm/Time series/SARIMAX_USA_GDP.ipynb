# SARIMAX

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
gdp_df = sm.datasets.macrodata.load_pandas().data
gdp_df
```

```python
gdp_df.realgdp.plot()
```

```python
gdp_train = gdp_df.iloc[:-20,2:8]
gdp_test = gdp_df.iloc[-20:,2:8]
```

```python
gdp_train
```

```python
realgdp_train = gdp_train.realgdp
features_train = gdp_train.iloc[:, 1:]
realgdp_test = gdp_test.realgdp
features_test = gdp_test.iloc[:,1:]
```

```python
features_train
```

```python
fitted_sarimax = SARIMAX(realgdp_train,
                         features_train,
                         order=(3,1,3),
                         seasonal_order=(0,0,0,4)).fit()
```

```python
predicted_gdp = fitted_sarimax.forecast(steps=20, exog=features_test)
predicted_gdp
```

```python
pd.concat([predicted_gdp, realgdp_test], axis=1).plot()
```

```python
fitted_sarimax.summary()
```

```python
fitted_sarimax.plot_diagnostics()
```
